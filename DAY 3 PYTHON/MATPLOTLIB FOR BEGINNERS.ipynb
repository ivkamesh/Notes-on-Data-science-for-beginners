## Object Oriented

### The main idea when using the formal object oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it

#### To begin we create a figure instance and then we add axes to that figure:

```python
#Create figure = empty canvas()

fig = plt.figure()

#Add set of axes to the figure

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'r')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Linear graph')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('MAIN PLOT')
axes2.set_title('SUB PLOT')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.2,0.4,0.3])
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)

#To fix the overlaps, I used the code line below
plt.tight_layout()
```

### Axes is just an array type, what this means is that it can be iterated through and indexed just the way any other array can, array operations can be performed on axes.

```python
fig,axes = plt.subplots(nrows=1, ncols=4)

# iterating through an axes array as with a normal array

for ax in axes:
    ax.plot(x,y)
    
    
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

# Indexing an axis and performing operations on them
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

### The same arguments can also be passed to layout managers such as the subplot function:

```python
fig, axes = plt.subplots(figsize=(10,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

# Special Plot Types

There are many specialized plots we can create, such as barplots, histograms, scatter plots, and much more. Most of these type of plots can be created even more efficiently using seaborn, a statistical plotting library for Python. But here are a few examples of these type of plots:

```python
plt.scatter(x,y)
```

```python
data = np.random.randint(1, 1000, 100)
plt.hist(data)
```

```python
data = np.random.randint(1, 1000, 100)

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)   #check out when vert and patch artist is false
```

```python
plt.hist(data, alpha=0.9, color="blue", edgecolor="black")
```

```python
labels =["Facebook", "Twitter", "Instagram", "Linkedin"]
size = [40, 25, 30, 5]
plt.pie(size, labels=labels, shadow=False, startangle=90, autopct="%1.i%%")
```

```python
plt.bar(labels, size, color="black", alpha=0.5)
```

```python
plt.barh(labels, size, color="black", alpha=0.5)
```
